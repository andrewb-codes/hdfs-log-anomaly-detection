## 2.2 Длина последовательности событий по классам

```python
def parse_features(s: str) -> list[str]:
    # "[E5,E22,E5,...]" -> ["E5", "E22", "E5", ...]
    if pd.isna(s):
        return []
    s = str(s).strip()
    if s.startswith('[') and s.endswith(']'):
        s = s[1:-1].strip()
    if not s:
        return []
    return [event_id.strip() for event_id in s.split(',') if event_id.strip()]


traces['sequence_len'] = traces['Features'].map(lambda s: len(parse_features(s)))
print(traces['sequence_len'].describe())
```

## 2.5 TimeInterval (csv + boxplots)

### Проверка согласованности TimeInterval и Features

```python
def parse_ti(s: str) -> list[float]:
    arr = s[1:-1].split(', ')
    return [float(x) for x in arr]

ti_len = traces['TimeInterval'].map(lambda s: len(parse_ti(s)))
print('TimeInterval len mismatch event sequence len count:', (ti_len != traces['sequence_len'] - 1).sum())
```

`TimeInterval` согласован с `Features`: для каждой последовательности есть интервал для каждой пары соседних событий, то есть `len(TimeInterval) = len(Features) - 1`.


### Статистики по TimeInterval для каждого трейса

```python
def ti_stats(s: str):
    a = np.array(parse_ti(s), dtype=float)
    return pd.Series({
        'ti_sum': a.sum(),
        'ti_mean': a.mean(),
        'ti_max': a.max(),
        'ti_std': a.std(ddof=0),
        'ti_zeros': (a==0).mean()
    })

ti_series = traces['TimeInterval'].apply(ti_stats)
ti_series = pd.concat([traces[['BlockId']], ti_series], axis=1).merge(labels[['BlockId', 'isAnomaly']], on='BlockId')

ti_lat_df = ti_series.merge(traces[['BlockId', 'Latency']], on='BlockId')

ti_lat_df[['BlockId', 'ti_sum', 'ti_mean', 'ti_max', 'ti_std', 'ti_zeros']].to_csv(
    OUT / 'timeinterval_stats_per_trace.csv', index=False)
ti_lat_df
```

`TimeInterval` согласован с `Latency`: сумма межсобытийных интервалов равна общей задержке трейса.

## 2.9 Переходы (пары последовательных событий)

### Топ 30 переходов (csv)

Смотрим самые частые пары последовательных событий

```python
def count_transition_pairs(df: pd.DataFrame) -> Counter:
    counts = Counter()
    for seq in df['Features'].map(parse_features):
        for a, b in zip(seq, seq[1:]):
            counts[(a, b)] += 1
    return counts


transitions_counter = count_transition_pairs(traces)
print('Unique transitions count:', len(transitions_counter))
```

```python
event_transitions_top30_df = pd.DataFrame([{
    'From': key[0],
    'To': key[1],
    'Count': value
} for key, value in transitions_counter.most_common(30)])

event_transitions_top30_df.to_csv(OUT / 'event_transitions_top30.csv', index=False)

print('Top 10 most frequent transitions:')
print(event_transitions_top30_df.head(10))
```

В топе типичные операционные циклы: E11->E9, E9->E11, а также повторы E26->E26, E23->E23, E21->E21, E5->E5.


### Топ 30 переходов типичных для каждого класса (csv)

Смотрим самые частые (по log_ratio) пары последовательных событий по долям для каждого класса

```python
table = (traces[['BlockId', 'Features']].merge(
    labels[['BlockId', 'isAnomaly']], on='BlockId'))

transitions_normal_counter = count_transition_pairs(table[table.isAnomaly == 0])
transitions_anomaly_counter = count_transition_pairs(table[table.isAnomaly == 1])

all_keys = set(list(transitions_normal_counter.keys()) + list(transitions_anomaly_counter.keys()))
rows = []
sum_norm = sum(transitions_normal_counter.values()) or 1
sum_anom = sum(transitions_anomaly_counter.values()) or 1

for k in all_keys:
    cN = transitions_normal_counter.get(k, 0)
    cA = transitions_anomaly_counter.get(k, 0)
    cT = cN + cA
    pN = cN / sum_norm
    pA = cA / sum_anom

    ratio = math.log((pA + 1e-12) / (pN + 1e-12))
    rows.append((k[0], k[1], cN, cA, cT, pN, pA, ratio))

event_transitions_df = pd.DataFrame(
    rows,
    columns=['prev', 'next', 'cnt_norm', 'cnt_anom', 'cnt_total', 'p_norm', 'p_anom', 'log_ratio']
)

df_filtered = event_transitions_df[event_transitions_df['cnt_total'] >= 100]
event_transitions_skew_anomaly_top30_df = df_filtered.sort_values('log_ratio', ascending=False).head(30)
event_transitions_skew_normal_top30_df = df_filtered.sort_values('log_ratio').head(30)

event_transitions_skew_anomaly_top30_df.to_csv(OUT / 'event_transitions_skew_anomaly_top30.csv', index=False)
event_transitions_skew_normal_top30_df.to_csv(OUT / 'event_transitions_skew_normal_top30.csv', index=False)
```

```python
print("Top 10 event transitions skew to normal:")
print(event_transitions_skew_normal_top30_df[['prev','next','cnt_anom','cnt_norm','p_anom','p_norm','log_ratio']].head(10))
```

```python
print("Top 10 event transitions skew to anomalies:")
print(event_transitions_skew_anomaly_top30_df[['prev','next','cnt_anom','cnt_norm','p_anom','p_norm','log_ratio']].head(10))
```

- `log_ratio > 0` -> переход диспропорционально чаще в аномалиях, `< 0` - чаще в норме.
- Среди топа склонных к норме - контекст финализации/учета. Эти переходы массово встречаются в норме и значительно реже в аномалиях.
- Среди топа склонных к аномалиям много переходов с `cnt_norm = 0`: в полном датасете они сотни раз происходят в аномалиях и ни разу в норме. Это сильный EDA-сигнал, но для моделирования такие выводы нужно пересчитывать только на train split.

Вывод: нормальные логи характеризуются устойчивыми сервисными цепочками (E26->E23, E9->E26, E22->E11 и т.д.), а аномалии - резкими переходами к ошибочным или нетипичным состояниям после приема/выделения блоков.


### Матрица переходов (heatmap)

Смотрим, насколько типичен для события `prev` каждый исход `next` и как это меняется в аномалиях. Строим матрицы условных вероятностей `P(next|prev)` отдельно для нормы и аномалий.

В отличие от топов выше, для heatmap используем vocabulary всех событий из справочника шаблонов, чтобы anomaly-only события не выпадали из матрицы.


```python
events_all_list = sorted(
    templates['EventId'].tolist(),
    key=lambda event_id: int(event_id[1:])
)
idx = {event_id: i for i, event_id in enumerate(events_all_list)}
```

```python
def build_transition_matrix(df: pd.DataFrame) -> tuple[np.ndarray, np.ndarray]:
    matrix = np.zeros((len(events_all_list), len(events_all_list)), dtype=float)
    row_counts = np.zeros(len(events_all_list), dtype=float)
    for seq in df['Features'].map(parse_features):
        for a, b in zip(seq, seq[1:]):
            if a in idx and b in idx:
                ia, ib = idx[a], idx[b]
                matrix[ia, ib] += 1.0
                row_counts[ia] += 1.0
    return matrix, row_counts


matrix_norm, rc_norm = build_transition_matrix(table[table.isAnomaly == 0])
matrix_anom, rc_anom = build_transition_matrix(table[table.isAnomaly == 1])

P_norm = (matrix_norm.T / (rc_norm + 1e-12)).T
P_anom = (matrix_anom.T / (rc_anom + 1e-12)).T
Delta = P_anom - P_norm

pd.DataFrame(P_norm, index=events_all_list, columns=events_all_list).to_csv(OUT / 'P_norm.csv')
pd.DataFrame(P_anom, index=events_all_list, columns=events_all_list).to_csv(OUT / 'P_anom.csv')
pd.DataFrame(Delta, index=events_all_list, columns=events_all_list).to_csv(OUT / 'P_delta.csv')
```

```python
def plot_heatmap(matrix: np.ndarray, title: str, file_name: str) -> None:
    plt.figure(figsize=(9, 7))
    plt.imshow(matrix, aspect='auto')
    plt.colorbar()
    plt.xticks(range(len(events_all_list)), events_all_list, rotation=90)
    plt.yticks(range(len(events_all_list)), events_all_list)
    plt.title(title)
    plt.tight_layout()
    plt.savefig(FIG / file_name, dpi=300)


plot_heatmap(P_norm, 'P(next | prev) - normal', 'transitions_norm_heatmap.png')
plot_heatmap(P_anom, 'P(next | prev) - anomaly', 'transitions_anomaly_heatmap.png')
plot_heatmap(Delta, 'Delta P (anom - norm)', 'transitions_delta_heatmap.png')
```

Строка - событие `prev` (что было до), столбец - событие `next` (что наступило).

Теплая клетка = для данного `prev` этот `next` встречается часто. Холодная клетка = для данного `prev` этот `next` случается редко.

`P_delta` показывает отличие аномального поведения от нормального: теплые клетки - переходы чаще в аномалиях, холодные - чаще в норме.


- Есть много общих переходов для аномалий и нормы: E11->E9, E9->E11 и повторы E21->E21, E23->E23, E26->E26 и др.
- В аномалиях резко чаще становятся E27->E23, E25->E5, E21->E20.
- В норме чаще, чем в аномалиях, встречаются E21->E21, E27->E4, E25->E18, E18->E5, E22->E11, E27->E26, E26->E26.
- Для дальнейших моделей матрицы переходов и списки anomaly-skewed событий нужно строить только по train split; в этом notebook они используются как EDA по полному датасету.
